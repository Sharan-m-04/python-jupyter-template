### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_sizes = [40, 25, 20, 15]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Line chart (First subplot)
axs[0, 0].plot(x, y, label='y = x²', color='blue', linewidth=2)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Plot 2: Bar chart (Second subplot)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Plot 3: Scatter plot (Third subplot)
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', s=100, edgecolor='black')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]),
                       textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9)

# Plot 4: Pie chart (Fourth subplot)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
