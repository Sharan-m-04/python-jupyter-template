### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = x**2  # y values are the square of x

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x²', color='blue', linewidth=2)

# Customize the plot
plt.title('Line Plot of y = x²', fontsize=14)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)

# Show the plot
plt.show()
```
